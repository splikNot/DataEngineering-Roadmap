### JSON

**JSON** (*JavaScript Object Notation*): é um formato leve de intercâmbio de dados. Ele é amplamente usado para transmitir dados entre sistemas, especialmente em aplicações web. Ele organiza os dados em chaves-valores e tem uma estrutura semelhante à objetos em javascript e dicionários em python.

```python
# Módulo JSON em python para trabalhar com arquivos do tipo
import json

# Dictionary to JSON string
data = {'name':'John', 'age':30}
json_data = json.dumps(data)

# JSON string to Python Dictionary
json_str = '{"name":"John", "age":30}'
data = json.loads(json_str)

# Python list to JSON array
data = [1, 2, 3] 
json_data = json.dumps(data)

# JSON array to Python list
json_str = '[1, 2, 3]'
data = json.loads(json_str)
```

* load - Lê os dados no arquivo json;
* dump: "Serializa" um objeto em python em json;
* serializar: Converter um objeto em python para json;
* desserializar: Operação inversa;
* read: Pegar o conteúdo de um arquivo em strings;
* write: Salva os dados em um arquivo

```python
import json

# Write Python dict to JSON file
data = {'name':'John'}
with open('data.json','w') as f:
  json.dump(data, f)
  
# Read JSON file to Python dict  
with open('data.json') as f:
  data = json.load(f)

# verify files exist
import os
file_exists = os.path.exists('data.json')
print("Verifying if data.json file exists:", file_exists)
```

### SQL

O **SQL** (*Structured Query Language*) é uma linguagem padrão para gerenciamento de bancos de dados relacionais. Ele é usado para criar, manipular, consultar e gerenciar os dados armazenados em sistemas de gerenciamento de banco de dados relacionais

*Termos-chave:*

* Query: Consultar e obter dados de um banco de dados;
    
* Select: Especifica colunas para retornar de um banco de dados;
    
* Where: Filtra as linhas de retornadas em umaa consulta (query);
    
* Wildcard: Caracteres especiais % e _ para encontrar padrões;
    
    - '*' : Realiza a operação para todos as linhas/campos do banco;
    - _s: Procura por qualquer caractere até que tenha a ocorrencia do caractere final;
        
* Order by: Ordena as colunas por uma ordem;
    
* Distinct: Remove linhas duplicadas de algum resultado;
    
* Table: Tabela que compoem o banco de dados organizados em linhas e colunas;
    
* Processo: Realizar alguma operação com seus dados.

## Web Scraping

Web Scraping é uma forma de acessar informações de websites mesmo quando não temos acesso a API delas.

*Termos-chave:*

* HTML Parsing: Análise e extração de dados HTML usando ferramentas como BeautifulSoup ou Python HMTLParser;
    
* API (Appplication Programmer Interface): Permite que programas conversem entre eles, geralmente através de arquivos do tipo JSON;
    
* Dados Desestruturados: São arquivos que não possuem um formato pré definido como os arquivos de HTML. Dessa forma os tornando difícil de se analisar;
    
* JSON (JavaScript Object Notation): Vilão da Classica saga Slasher, Sexta-Feira 13. Por ser um formado onde os dados estão estruturados é bem mais simples para programas interagir e analisar;
    
* Normalizar: Pegar dados bagunçados e desestruturados e arruma-los para torna-los consistentes;
    
* Dados Persistentes (Não entendi o pq dessa nomenclatura): Salvar dados extraídos de algum banco de dados em um arquivo para ser usado depois;
    
* XML(Extensible Markup Language): Outro formato de estrutura de dados semelhante ao JSON (Perderam a chance de chama-lo de Freddy);

* CSV (Comma Separated Values): Esse já conheço bem, um formato de tabular só que em um arquivo de texto simples;

* Engenharia de Dados: Manipular, limpar e transformar dados crus em um formato mais legivel e útil.
    

```python
from html.parser import HTMLParser

class ParseResults(HTMLParser): #HTML Parser funciona como uma classe

    def __init__(self): #Iniciando os atributos
        super().__init__() #Como iniciamos uma classe que herdará as caracteristicas de outras. Fazemos o super()
                            #para fazer o init da classe pai/mãe (qual é o genero de uma classe?)
        self.recording = False 
        self.title = []

    def handle_starttag(self, tag, attrs): #Metodo para procurarmos as tags 
        if tag == "title":
            self.recording = True
        else:
            self.recording = False
            
    def handle_data(self, data): #Método que irá realizar ação conforme for encontrado o que buscamos
        if self.recording:
            self.title.append(data)

#Não tava conseguindo encontrar o caminho correto do arquivo, consegui resolver com
#print(os.getcwd())

p = ParseResults()
with open("/home/cyanide/Documentos/Cursos/Python-Bash-and-SQL-Essencials-for-Data-Enginering/C3-Scripting_with_Python_and_SQL/1992_Atlh_Jumps.txt") as _f:
    p.feed(_f.read())

print(p.title)
```

### Efficient Scraping Techniques

(Texto diretamente do curso que eu achei bem relevante, não vou traduzir pra ninguém pensar que foi eu quem escreveu)

**Efficiency**

Usually, a spider will have a set of URLs that it will need to request when scrapping. However, there are several problems with this approach when starting a scrapping project to gather data. Let's go into some of the details regarding parsing live websites and requesting data each time.

**First iterations are always hard**

In the current repo, I already solved how to get to the data. However, when you start working with a live website, you will have to take the time to find how to get to the data. Then, every time you "try it out" again, you need to make a live request to the website and download its contents.

This is not efficient, and it can be time-consuming. It also puts unnecessary strain on the webserver hosting the website. Being "nice" when parsing requires being aware that websites are resources that take money and effort to be available. Preventing an unnecessary burden to these websites is suitable for webservers and efficient for you since it is faster to read from a file. Every single time you need to start a scrapy shell again, it will be an order of magnitude faster.

**Testing**

Websites change, and you might be getting some changes that are disruptive to your parsing. You can tweak the HTML and then use that tweaked file to run the spider against it by downloading the contents directly. This allows you to build a robust spider that can handle oddities on website content. Changing HTML on the fly would be very difficult to get right instead of just changing the file directly.

**Speed**

I've already mentioned that parsing a local file is an order of magnitude faster. There are always problems with scrapping, like network errors, downtime on webservers, or even memory errors when parsing lots of data. Imagine you have to parse several thousand pages and get into an unrecoverable error in the middle. It would be time-consuming to have to start from the very beginning.

**Updating the spider**

If the requirements of the gathered data change, then you don't want to have to re-parse all those pages again. Of course, this depends if the data keeps changing. When updating the data requirements, you can use the same pages available on disk to do this.
